```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
# kfold 설정
kfold = StratifiedKFold(n_splits=5)
```

## RF gridsearchCV (0.78945)

```python
#RandomForestClassifier

rf_param_grid = {"max_depth": [None],
              "min_samples_split": [3, 5, 10],
              "min_samples_leaf": [1, 3, 5],
              "n_estimators" :[100,300,400]}

gsRF = GridSearchCV(rf, param_grid=rf_param_grid, cv=kfold, scoring="f1", n_jobs=4, verbose=1)

gsRF.fit(tfidf_xtrain, y_train)

RF_best = gsRF.best_estimator_

#Best score
gsRF.best_score_
```

```python
#Best parameter
gsRF.best_params_
```

```python
RF_best.fit(tfidf_xtrain, y_train)
pred_gsRF = RF_best.predict(tfidf_xtest)
```
